```python
import scrapy
from ScrapyElasticSearch import ElasticsearchPipeline


class MySpider(scrapy.Spider):
    name = 'my_spider'
    start_urls = ['https://rule.handong.edu']  # 크롤링할 URL 목록

    custom_settings = {
        'FEED_EXPORT_ENCODING': 'euc-kr',
        'ELASTICSEARCH_SERVERS': ['https://9a8061ea253345d9bdc0ca771dcb49c7.ap-northeast-2.aws.elastic-cloud.com:443'],
        'ELASTICSEARCH_INDEX': "hgu_rules",
        'ELASTICSEARCH_TYPE': '_doc',
        'ELASTICSEARCH_UNIQ_KEY': 'url',  # 데이터의 고유 키
        'ITEM_PIPELINES': {
            'scrapy_elasticsearch.ElasticSearchPipeline': 500,
        }
    }

    def parse(self, response):
        yield {
            'url': response.url,
            'title': response.css('title::text').get(),
            'body': response.css('body').get(),
        }
```
